```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Définir l’URL de la page Yahoo Finance (Tesla)
url_tesla = "https://finance.yahoo.com/quote/TSLA/financials"

# Faire une requête GET pour récupérer la page
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url_tesla, headers=headers)

# Vérifier si la requête a réussi
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    print(soup.prettify())  # Affiche le HTML de la page pour analyser
else:
    print("Erreur lors de la récupération de la page :", response.status_code)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import yfinance as yf
```

```python
def make_graph(data, title):
    """
    Function to plot a graph of the closing prices of a stock.
    
   Parameters: - data: DataFrame containing stock market data (e.g. tesla_data) 
   
   - title: Title of the graph
    """
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label="Closing Price", color='blue', linewidth=2)

    # Add a title and labels
    plt.title(title, fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Closing Price ($)")
    plt.legend()
    plt.grid()

    # View Chart
    plt.show()
```

## Question 5: Plot Tesla Stock Graph

```python
# Reset the index so that the "Date" column is usable
tesla_data_reset = tesla_data.reset_index()

# Plot the Tesla stock chart
make_graph(tesla_data_reset, "Tesla stock price action (TSLA) Evolution")
```
